```python
import re # Validado con regular expresión

def validar_nif_nie(nif_nie):
    #Expresiones regulares para NIF y NIE

    #Verificamos si el formato coincide con un NIF
    if re.match(r'^\d{8}[A-Z]$', nif_nie):
        validar = "TRWAGMYFPDXBNJZSQVHLCKE"
      #Calculamos el valor para la letra de control

        #Comprobamos si la letra de control coincide con la calculada
        return nif_nie[-1] == validar[int(nif_nie[:-1]) % 23]

    # Verificar si el formato coincide con un NIE
    elif re.match(r'^[XYZxyz]\d{7}[A-Za-z]$', nif_nie):
        #Convertimos la letra en el número correspondiente
        if nif_nie[0].lower() == "x":
            nie_num = "0" + nif_nie[1:]
        elif nif_nie[0].lower() == "y":
            nie_num = "1" + nif_nie[1:]
        elif nif_nie[0].lower() == "z":
            nie_num = "2" + nif_nie[1:]
        else: return False

        validar = "TRWAGMYFPDXBNJZSQVHLCKE"
        numeros = nie_num[:-1]  #Obtenemos los números
        resultado = int(numeros) % 23 #Calculamos el valor para la letra de control

        #Convertimos la letra en el número correspondiente
        if nie_num[-1] == validar[resultado]:
            return True
        else:
            return False
    else:
        return print("El formato no es válido")
```

```python
validar_nif_nie("52505013T") #Nie invalido
```

```python
validar_nif_nie("Y2345678Z") #NIE válido
```

```python
validar_nif_nie("71164300T") #NIF válido
```

```python
validar_nif_nie("54410369F") #NIF Inváido
```
