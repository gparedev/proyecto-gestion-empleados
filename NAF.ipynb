Validación NAF

```python
import re

def validacionNAF(naf):
    validarNaf = r'^\d{12}$'

    if re.match(validarNaf, naf): #Validamos el formato de NAF
        a, b, c = map(int, [naf[0:2], naf[2:10], naf[10:]]) # Extraemos las tres partes
        # Comprobamos si b es menor que 10000000
        if b < 10000000:
            #Comprobamos si coincide con c
            codigo = (b + a * 10000000) % 97
            if c == codigo:
                return True
            else:
                return False
        else:
            #Verificamos si coincide con c.
            codigo = int(str(a) + str(b)) % 97
            if c == codigo:
                return True
            else:
                return False
    else:
        print("El formato no es válido")


```
