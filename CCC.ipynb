Validar dígito de control

```python
def digitoControl(numeros):
    multiplica = [1, 2, 4, 8, 5, 10, 9, 7, 3, 6]
    rango = len(multiplica)
    # Verificamos si la longitud de la cadena es diferente de 10
    if len(numeros) != 10:
        multiplica.remove(1)  # Si es así eliminamos los dos primeros valores
        multiplica.remove(2)
        rango -= 2  # Reducioms el rango de iteración a 8
    suma = sum(map(lambda x: int(numeros[x]) * multiplica[x], range(rango)))
    if 11 - (suma % 11) == 10:
        return 1
    if 11 - (suma % 11) == 11:
        return 0

    return 11 - (suma % 11)
```

Validación de CCC

```python
def validarCCC(ccc):
    if str(digitoControl(ccc[0:8])) + str(digitoControl(ccc[10:])) == ccc[8:10]:
        return True
    else:
        return False
```
